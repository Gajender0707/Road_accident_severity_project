```python
## Now handle the missing values as there are categorical features mostly then we don't need to worry about like outliers , etc.
# we just need to fill the null values using the mode 
```

```python
## filling the null values
for i in data.columns:
    data[i]=data[i].replace("na",np.nan)
    if data[i].isnull().sum()>0:
        data[i]=data[i].fillna(data[i].mode()[0])
```

```python
data.isnull().sum()
```

```python
for i in data.columns:
    if data[i].dtype in ["object"]:
        print(data[i].value_counts())
        print(">>>>>>>"*30)
```

## Observation
1. if the anyone feature  is  occupining more then 85% by only one category means that is no much variations in the feature to provide the enough information for the mode

```python
def show_feature_information(data):
    useless_col=[]
    for col in data.columns:
        if col=="Time":
            continue
        print(col)
        info={}
        if data[col].dtype in ["object"]:
            per=dict(data[col].value_counts())
            for i,j in per.items():
                info[i]=round(j/len(data)*100,2)
                print(f"{i}: {round(j/len(data)*100,2)}%")
        for i,j in info.items():
            if j>85:
                print(f"{col} must be exit")
                useless_col.append(col)
        print(">>>>>>>"*30)
    return useless_col
```

```python
useless_col=show_feature_information(data)
```

```python
useless_col
```

```python
data=data.drop(useless_col,axis=1)
```
