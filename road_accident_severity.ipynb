```python
def show_feature_information(data):
    useless_col=[]
    for col in data.columns:
        if col=="Time":
            continue
        print(col)
        info={}
        if data[col].dtype in ["object"]:
            per=dict(data[col].value_counts())
            for i,j in per.items():
                info[i]=round(j/len(data)*100,2)
                print(f"{i}: {round(j/len(data)*100,2)}%")
        for i,j in info.items():
            if j>85:
                print(f"{col} must be exit")
                useless_col.append(col)
        print(">>>>>>>"*30)
    return useless_col
```

## Reducing Category
as there are many such category which happens in very less frequency these are not providing very much value thus these can also be added in the other category

```python
## first of all the list the columns in which need to reduce the category.
```

```python
for i in data.columns:
    if data[i].dtype in ["object"]:
        print(data[i].value_counts())
        print(">>>>>>>"*30)
```

```python
cat_reduce_col=[]
for i in data.columns:
    if data[i].dtype=="O":
        val_count=data[i].value_counts().values
        for j in val_count:
            if j<=200:
                cat_reduce_col.append(i)
```

```python
cat_reduce_col_list=np.unique(cat_reduce_col)
```

```python
len(np.unique(cat_reduce_col))
```

```python
df=data.copy()
```

```python
df.head()
```

```python
for col in cat_reduce_col_list:
    if col!="Accident_severity":
        col_val_count=df[col].value_counts()
        col_fair_values=col_val_count[col_val_count>150].index
        # print(col_fair_values)
        # print(">>>>"*30)
        df[col]=df[col].apply(lambda x: x if x in col_fair_values else "Other")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier,
    BaggingClassifier,
    VotingClassifier,
    StackingClassifier
)
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Naive Bayes": GaussianNB(),
    "MLP": MLPClassifier()
}
```

```python
from sklearn.model_selection import train_test_split

# Let's say your features are in X and labels are in y
# X = df.drop("target_column", axis=1)
# y = df["target_column"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y,              # Features and target
    test_size=0.2,     # 20% for testing, 80% for training
    random_state=42,   # For reproducibility
    stratify=y         # Optional: ensures balanced target in both sets
)
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"{name} Accuracy: {model.score(X_test, y_test):.2f}")
```
