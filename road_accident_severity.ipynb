```python
fig,axes=plt.subplots(9,2,figsize=(30,30))
for i,feature in enumerate(cat_features):
    r,c=divmod(i,2)
    sns.countplot(x=data[feature],hue=data["Accident_severity"],ax=axes[r,c])
```

```python
new_cat=cat_features[:6]
new_cat
```

```python
fig,axes=plt.subplots(3,2,figsize=(20,20))
for i,feature_name,in enumerate(new_cat):
    r,c=divmod(i,2)
    label=data[feature_name].value_counts().index
    val=data[feature_name].value_counts().values
    axes[r][c].pie(val,labels=label,autopct="%1.1f%%")
    axes[r][c].set_title(f"{feature_name.capitalize()} distribution")
```
